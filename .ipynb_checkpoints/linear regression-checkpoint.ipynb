```python
Линейная регрессия — это метод анализа данных, который предсказывает ценность неизвестных данных с помощью другого связанного и известного 
значения данных. Он математически моделирует неизвестную или зависимую переменную и известную или независимую переменную в виде линейного уравнения.
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
df = pd.read_csv('Student_Performance_.csv')
df.head()
print(f'rows: {df.shape[0]}')
print(f'features: {df.shape[1]}')
x = df["Previous Scores"].values.reshape(-1,1)
y = df["Performance Index"].values.reshape(-1,1)
df
df.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
plt.figure(figsize=(5, 4))
sns.heatmap(pd.DataFrame(X_scaled).corr(), annot=True, cmap='coolwarm')
plt.title("Тепловая карта корреляции признаков")
plt.show()
```

```python
X_train,X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=987)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
train_prediction = model.predict(X_train);
```

```python
print("Результат на тестовом наборе:", model.score(X_test, y_test))
print("Результат на обучающем наборе:", model.score(X_train, y_train))
```

```python
plt.scatter(X_train,y_train,color = "green")
plt.plot(X_train,train_prediction,color = "red")
plt.title('Dependency Student Index and Previous Scores')
plt.xlabel('Previous Scores')
plt.ylabel('Performance Index')
plt.show()
```
