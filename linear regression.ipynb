```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Student_Performance_.csv')
df.head()
```

```python
print(f'rows: {df.shape[0]}')
print(f'features: {df.shape[1]}')
```

```python
x = df["Previous Scores"].values.reshape(-1,1)
y = df["Performance Index"].values.reshape(-1,1)
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=987)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
print(model.score(x_test, y_test))
```

```python
print(model.score(x_train, y_train))
```

```python
train_prediction = model.predict(x_train);
```

```python
plt.scatter(x_train,y_train,color = "green")
plt.plot(x_train,train_prediction,color = "red")
plt.title('Dependency Student Index and Previous Scores')
plt.xlabel('Previous Scores')
plt.ylabel('Performance Index')
plt.show()
```
