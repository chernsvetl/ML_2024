```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score, roc_auc_score, accuracy_score, precision_score, recall_score, mean_absolute_error, mean_squared_error
```

```python
data = datasets.load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
X = data.data
y = data.target
df.head()
df
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(pd.DataFrame(X_scaled).corr(), annot=True, cmap='coolwarm')
plt.title("Тепловая карта корреляции признаков")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, random_state=987)
```

```python
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
```

```python
y_pred = nb_classifier.predict(X_test)
y_pred_proba = nb_classifier.predict_proba(X_test)[:, 1]
```

```python
cf_matrix = confusion_matrix(y_test, y_pred, labels=np.unique(y))
print("Confusion Matrix:")
print(cf_matrix)

plt.figure(figsize=(3, 3))
ConfusionMatrixDisplay(cf_matrix).plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
print(accuracy_score(y_test, y_pred))
print(precision_score(y_test, y_pred, average='macro'))
print(recall_score(y_test, y_pred, average='macro'))
print(f1_score(y_test, y_pred, average='macro'))
print(roc_auc_score(y_test, y_pred_proba))
```
